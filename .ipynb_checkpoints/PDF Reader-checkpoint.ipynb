```python
import pandas as pd
import tabula as tb
```

```python
filePath = "sample_pdf.pdf"
```

```python
tableList = tb.read_pdf(filePath, pages="all", silent=True, pandas_options={'header': None} )
```

```python
tabLength = len(tableList)
print(tabLength)

# append
df = tableList[0].append(tableList[1:tabLength])

# reindex 
df.reset_index(inplace = True,drop = True)
df
```

```python
df.rename(columns={0: 'Date', 1: 'Narration', 2:'Chq', 3:'ValueDt', 4:'WithDrawalAmt', 5:'DepositAmt', 6:'ClosingBalance'}, inplace=True)
data =df
```

```python
# Drop the first record
data = data.drop(0)
data.head()
```

```python
len(data)
```

```python
data[2:3]
```

```python
for index, row in data.iterrows():
    
    # Skip the first record
    if index == 0:
        continue
    
    x = row.Date
    
    if pd.isnull(x): 
        print("Index -", index )   
        
        # Append Previous record of Narration + current Narration
        narMsg= data[index:index]['Narration'] + row.Narration
        
        # Update Previous Record
        data[index:index]['Narration'] = narMsg
        
        print(narMsg)
```

```python
# drop record if date filed is null 
newdf = data.dropna(subset=['Date'])
newdf
```
